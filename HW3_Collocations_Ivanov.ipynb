# Домашнее задание №3. Collocations


Анализ коллокаций на выборке из корпуса судебных разбирательств с помощью nltk.

Размер корпуса: 5083 слова. 

Использовано две метрики: Pointwise Mutual Information (PMI) и Log-Likelihood. 

## Золотой стандарт

-  ('ВЫНЕСТИ' ,'РЕШЕНИЕ'),
-  ('ИСПОЛНЯТЬ' ,'РЕШЕНИЕ'),
-  ('ПРИЗНАТЬ', 'ПРАВОТА'),
-  ('ПРИЗНАТЬ', 'ВИНОВНАЯ'),
-  ('ВОЗОБНОВИТЬ' ,'СЛУШАНИЕ'),
-  ('ПРИНЯТЬ', 'РЕШЕНИЕ'),
-  ('УДОВЛЕТВОРИТЬ', 'ИСК'),
-  ('ВЫНЕСТИ' ,'ПОСТАНОВЛЕНИЕ'),
-  ('ДОКАЗАТЬ' ,'ФАКТ'),
-  ('ИЗБРАТЬ' ,'МЕРА')

```python
import nltk
from nltk.collocations import *
bigram_measures = nltk.collocations.BigramAssocMeasures()
```

```python
lines = open('court-V-N.csv', 'r', encoding='utf-8').read().split('\n')

bigram = []
for l in lines:
    line_splitted = l.split(',')[1:3]
    for w in line_splitted:
        bigram.append(w)
```

```python
finder = BigramCollocationFinder.from_words(bigram)
```

## С помощью Pointwise Mutual Information (PMI) 

```python
finder.apply_freq_filter(7) #частота встречаемости не менее 7 раз
PMI = finder.nbest(bigram_measures.pmi, 20) # из списка выбираем 10 конструкций V + Obj
PMI
```

## С помощью Log-Likelihood

```python
LL = finder.nbest(bigram_measures.likelihood_ratio, 25) # из списка выбираем 10 конструкций V + Obj
LL
```

```python
from nltk.metrics.spearman import *

# список топ-коллокаций V + Obj, полученный Pointwise Mutual Information
PMI = [('ВЫДАТЬ', 'САНКЦИЯ'),
 ('НАЛОЖИТЬ', 'АРЕСТ'),
 ('САНКЦИОНИРОВАТЬ', 'АРЕСТ'),
 ('ОТКАЗАТЬ', 'УДОВЛЕТВОРЕНИЕ'),
 ('ПРИЗНАТЬ', 'ВИНОВНАЯ'),
 ('ПРИЗНАТЬ', 'ПРАВОТА'),
 ('УДОВЛЕТВОРИТЬ', 'ХОДАТАЙСТВО'),
 ('ПРИНЯТЬ', 'РЕШЕНИЕ'),
 ('РАССМОТРЕТЬ', 'ХОДАТАЙСТВО'),
 ('УДОВЛЕТВОРИТЬ', 'ИСК')]

# список топ-коллокаций V + Obj, полученный Log-Likelihood
LL = [('ВЫДАТЬ', 'САНКЦИЯ'),
 ('ПРИНЯТЬ', 'РЕШЕНИЕ'),
 ('УДОВЛЕТВОРИТЬ', 'ХОДАТАЙСТВО'),
 ('УДОВЛЕТВОРИТЬ', 'ИСК'),
 ('НАЛОЖИТЬ', 'АРЕСТ'),
 ('ПРИЗНАТЬ', 'ВИНОВНАЯ'),
 ('САНКЦИОНИРОВАТЬ', 'АРЕСТ'),
 ('ОТКАЗАТЬ', 'УДОВЛЕТВОРЕНИЕ'),
 ('ПРИЗНАТЬ', 'ПРАВОТА'),
 ('ВЫНЕСТИ', 'РЕШЕНИЕ')]

# Золотой стандарт
GS = [('ВЫНЕСТИ' ,'РЕШЕНИЕ'),
 ('ИСПОЛНЯТЬ' ,'РЕШЕНИЕ'),
 ('ПРИЗНАТЬ', 'ПРАВОТА'),
 ('ПРИЗНАТЬ', 'ВИНОВНАЯ'),
 ('ВОЗОБНОВИТЬ' ,'СЛУШАНИЕ'),
 ('ПРИНЯТЬ', 'РЕШЕНИЕ'),
 ('УДОВЛЕТВОРИТЬ', 'ИСК'),
 ('ВЫНЕСТИ' ,'ПОСТАНОВЛЕНИЕ'),
 ('ДОКАЗАТЬ' ,'ФАКТ'),
 ('ИЗБРАТЬ' ,'МЕРА')]
```

```python
# сравним результаты PMI и Log-Likelihood между собой
print('%0.1f' % spearman_correlation(ranks_from_sequence(PMI), ranks_from_sequence(LL)))
```

```python
# сравним результаты PMI с Золотым стандартом
print('%0.1f' % spearman_correlation(ranks_from_sequence(PMI), ranks_from_sequence(GS)))
```

```python
# сравним результаты Log-Likelihood с Золотым стандартом
print('%0.1f' % spearman_correlation(ranks_from_sequence(LL), ranks_from_sequence(GS)))
```
